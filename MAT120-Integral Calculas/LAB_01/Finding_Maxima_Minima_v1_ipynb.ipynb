```python
from sympy import*
import numpy as np
import matplotlib.pyplot as plt
import math
from numpy import log as ln
```

```python
x=symbols('x')
y= 9 * x ** 3 - 7*x**2 +3*x + 10
lv= -5
uv = 6
dy=diff(y,x)
dy1=diff(y,x,2)

print(dy)
print(dy1)
print("Minima",y.subs(x,lv))
print("Maxima",y.subs(x,uv))

y=lambdify(x,y)
p=np.arange(lv,uv)
a=y(p)
print(a)

plt.plot(p,a)
plt.title('Shape')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```
